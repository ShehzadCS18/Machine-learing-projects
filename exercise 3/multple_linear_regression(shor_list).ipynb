```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X=np.loadtxt('ex3x.dat')
Y=np.loadtxt('ex3y.dat')
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scalar=StandardScaler()
X3=scalar.fit_transform(X)
```

```python
def model(x,y,lr,iteration):
    m=y.size
    theta=np.zeros((x.shape[1]))
    los=[]
    for epoch in range(iteration):
        y_pred=np.dot(x,theta)
        loss=(1/(2*m))*np.sum(np.square(y_pred-y))
        dervative_theta=(1/m)*np.dot(x.T,y_pred-y)
        theta=theta-np.dot(lr,dervative_theta)
        los.append(loss)
        if epoch % 1==0:
            print(f'epoch{epoch+1}: loss={loss:.3f}')
    return theta,los
```

```python
iteration=300
lr=0.01
theta,los=model(X3,Y,lr=lr,iteration=iteration)
```

```python
plt.plot(los,'b')
plt.show()
```

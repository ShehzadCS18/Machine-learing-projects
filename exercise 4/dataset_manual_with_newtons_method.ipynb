```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=np.loadtxt('ex4x.dat')
y=np.loadtxt('ex4y.dat')
```

```python
ones=np.ones([x.shape[0],1])
x=np.concatenate((ones,x),axis=1)
```

```python
x.shape
```

```python
w=np.zeros((x.shape[1]))
```

```python
def prediction_sigmiod(x):
    return 1.0/(1 + np.exp(-(np.dot(w,x.T))))
```

```python
def loss2(m,y,y_predicted):
    return (-(1/m)*np.sum( (y*np.log(y_predicted)) + (1-y)*np.log(1-y_predicted)))
```

```python
def gradient2(m,x,y,y_predicted):
    dw = (1/m)*np.dot(x.T, (y_predicted - y))
    return dw
```

```python
def hassian(m,x,y_predicted):
    H=(1/m)*(np.dot(y_predicted,(1-y_predicted))*(x))
    return H
```

```python
def hassian2(m,x,y_predicted):
    H=(1/m)*(np.sum((y_predicted*(1-y_predicted))*(np.dot(x,x.T))))
    return H
```

```python
def hassian3(m,x,y_predicted):
    H=(1/m)*(np.sum(np.dot(y_predicted,(1-y_predicted))*(np.dot(x,x.T))))
    return np.array(H)
```

```python
m=x.shape[0]
```

```python
# now training loop
iteration=500
lr=0.0001
los=[]
for epoch in range(iteration):
    y_pred=prediction_sigmiod(x)
    #l=loss(y,y_pred)
    l=loss2(m,y,y_pred)
    los.append(l)
    dw=gradient2(m,x,y,y_pred)
    h=hassian2(m,x,y_pred)
    #w=w-lr*dw
    #h_inverse=np.linalg.inv(h)
    #w=w-np.dot(h,dw)
    w=w- (dw/h)
    #w=w- (h/dw)
    
    if epoch % 1==0:
        print(f'epoch{epoch+1}:loss={l:.4f}')
```
